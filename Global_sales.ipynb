**Create a new column for Technology Category**

```python
Tech = []
for i in df['Category']:
    if i == 'Technology':
        Tech.append(i)
    else:
        Tech.append(None)
df['Tech'] = Tech
df
```

**Create a new column for Furniture Category**

```python
Furniture = []
for i in df['Category']:
    if i == 'Furniture':
        Furniture.append(i)
    else:
        Furniture.append(None)
df['Furniture'] = Furniture
df
```

**Create a new column for Office Supply Category**

```python
Office = []
for i in df['Category']:
    if i == 'Office Supplies':
        Office.append(i)
    else:
        Office.append(None)
df['Office'] = Office
df
```

**Create a new column for Technology sales**

```python
new_sales = [] #Create an empty list 
sales = df[df['Category'] == 'Technology'] #Direct the elements (Technology) in the Category series to the list (Sales)
```

```python
for index, row in sales.iterrows(): #Iterate through all the elements (Technology) in the Category Column
    new_sales.append(row['Sales']) #move all the elements in Sales that has thesame index as the element (Technology) in the Category column to the list created initially 
```

```python
_new_sales = [] #Create an empty list
for i in df['Sales']: #Iterate through the Sales series
    if i in new_sales: #If an element found in the sales series is in the new_sales list...
        _new_sales.append(i) #move the element to the empty list
    else:
        _new_sales.append(None) 
```

```python
df['tech_sales'] = _new_sales #Create the technology sales column
```

```python
df
```

**Create a new column for Furniture sales**

```python
new_sales1 = []
sales1 = df[df['Category'] == 'Furniture']
```

```python
for index, row in sales1.iterrows():
    new_sales1.append(row['Sales'])
```

```python
_new_sales1 = []
for i in df['Sales']:
    if i in new_sales1:
        _new_sales1.append(i)
    else:
        _new_sales1.append(None)
```

```python
df['Furniture_sales'] = _new_sales1
```

```python
df
```

**Create a new column for Office Suply sales**

```python
new_sales2 = []
sales2 = df[df['Category'] == 'Office Supplies']
```

```python
for index, row in sales2.iterrows():
    new_sales2.append(row['Sales'])
```

```python
_new_sales2 = []
for i in df['Sales']:
    if i in new_sales2:
        _new_sales2.append(i)
    else:
        _new_sales2.append(None)
```

```python
df['Office_Sales'] = _new_sales2
```

```python
df
```
